## Pandas style 1

Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.




```python
import numpy as np
```

```python
df1 = pd.DataFrame(np.random.rand(10,4))
df1
```

```python
df1.loc[2,1] = pd.NA
df1.loc[7,3] = pd.NA
df1
```

```python
df1.style.map(lambda v: 'background-color: green' if pd.isna(v) else None)
```

## Pandas style 2

Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframe’s specific columns with different colors.




```python
colors = ["green","blue","black","red"]
df1.style.map_index(lambda col : f'background-color : {colors[col]}',axis = 1)
```

```python
colors = ["green","blue","black","red"]
df1.style.apply(lambda x : [f'background-color : {colors[col]}' for col in x.index],axis = 1)
```
