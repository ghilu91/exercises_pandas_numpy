#### Select companies and incidents_85_99 with incidents_85_99 is less than 10

```python
mask = df['incidents_85_99'] < 10
df2 = df[mask][['airline','incidents_85_99']]
df2
```

#### Group companies by first letter(lower) and take the mean incidents_85_99 and make it a dataframe again 

```python
df2['airline_first_letter'] = df2['airline'].str[0] 
df3 = df2[['airline_first_letter','incidents_85_99']].groupby('airline_first_letter').mean()
```

```python

df3 = pd.DataFrame(df3, columns= ['incidents_85_99'])
df3.reset_index()
```
